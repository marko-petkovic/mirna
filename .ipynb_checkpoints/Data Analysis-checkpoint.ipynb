```python
import sys
#sys.path.insert(0,'/content/drive/MyDrive/Marko/master')
sys.path.insert(0, link)
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import OneHotEncoder

from tqdm import tqdm
from tqdm import trange

import datetime

from google.colab import drive
drive.mount('/content/drive')
```

```python
rna_combs = {}
```

```python
images = np.load(f'{link}/modmirbase_train_images.npz')['arr_0']/255
```

```python
def get_color(pixel):
    """
    returns the encoded value for a pixel
    """
    if (pixel == np.array([0,0,0])).all():  
        return 'Z' # black
    elif (pixel == np.array([1,0,0])).all():  
        return 'G' # red
    elif (pixel == np.array([0,0,1])).all():  
        return 'C' # blue
    elif (pixel == np.array([0,1,0])).all():  
        return 'U' # green
    elif (pixel == np.array([1,1,0])).all():  
        return 'A' # yellow
    else:
        print("Something wrong!")
```

```python
names =  np.load(f'{link}/modmirbase_train_names.npz')['arr_0']
names = [i.decode('utf-8') for i in names]
```

```python
labels = np.load(f'{link}/modmirbase_train_labels.npz')['arr_0']
```

```python
if labels[0]==1:
    print('ja')
```

```python
diff_len = []
double_black = []
for i in range(images.shape[0]):
    if i%1000==0:
        print(i)
    for j in range(images.shape[2]):
        if (images[i,12,j] == np.array([1,1,1])).all():
            break
        cols = get_color(images[i,12,j])+get_color(images[i,13,j])
        cols = "".join(sorted(cols))
        if cols in rna_combs:
            rna_combs[cols][0]+=1
            
        else:
            rna_combs[cols]=[1,0,0]
        
        if 'random' in names[i]:
            rna_combs[cols][1]+=1

        if labels[i] == 1:
            rna_combs[cols][2]+=1

        if cols == "ZZ":
            double_black.append(i)
        len1 = 0
        # loop until white pixel
        while not (images[i,12-len1,j] == np.array([1.,1.,1.])).all():
            len1 += 1
            if 13-len1 == 0:
                break
        
        len2 = 0
        while not (images[i,13+len2,j] == np.array([1.,1.,1.])).all():
            len2 += 1
            if 13+len2 == 25:
                break
        # if lenghts differ
        if len1 != len2 and len1+len2 != 25 and 'Z' not in cols:
            diff_len.append((i,(len1, len2), 1*('random' in names[i])))

double_black = list(set(double_black))
```

```python
rna_combs, len(diff_len)
```

```python
zp = sorted(zip([i for i in rna_combs], [rna_combs[i][0] for i in rna_combs], [rna_combs[i][2] for i in rna_combs]), key = lambda x:x[0])
```

```python
plt.figure(figsize=(30,10))
pl = plt.bar([i[0] for i in zp],[i[1] for i in zp], label = 'non miRNA')

pl2 = plt.bar([i[0] for i in zp],[i[2] for i in zp], label = 'miRNA')
for bar in pl:
    plt.annotate(bar.get_height(), 
                 xy=(bar.get_x()+0.0, bar.get_height()+10), 
                     fontsize=20)
plt.xticks(fontsize=20)
plt.title('Amount of occurences per nucleotide combination', fontsize=40)
plt.legend()
```

```python
x = 0
i = 0
old = []
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(120,30))

while x < 9:
    if diff_len[i][0] not in old:
        ax[x%3, x//3].imshow(images[diff_len[i][0]])
        x+=1
        old.append(diff_len[i][0])
    i+=1
    
  
```
